```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#cargar archivo
from google.colab import files
files.upload()
```

```python
df =pd.read_excel('Datos de Facturaci√≥n.xlsx')
df.head()
```

```python
#cambiar de str a float
df["CVE_CLPV"] = df["CVE_CLPV"].astype(float)
```

```python
df["CVE_CLPV"] = df["CVE_CLPV"].replace({'':'0'}, regex = False)
```

```python
df1 = df.copy()
df1.head()
```

```python
df1.info()
```

```python
# CVE_CLPV, valores: 1000 a 2000
filtro1 =df1[(df1["CVE_CLPV"] >=1000) & (df1["CVE_CLPV"] <=2000)]
filtro1
```

```python
# CVE_VEND, todas las claves excepto "4" y "5"
filtro2 = df1[~(df1['CVE_VEND'] == 4) & ~(df1["CVE_VEND"] == 5)]
filtro2
```

```python
# 3- FECHA_ENT - fechas de "28/02/2022"
df1['FECHA_ENT']=df1['FECHA_ENT'].astype(str)

filtro3 = df1[df1['FECHA_ENT'] == "2022-02-28 00:00:00"]
filtro3
```

```python
#4- CAN_TOT -Cantidades menores a 5951.7 o STATUS("E")
filtro4 = df1[(df1['CAN_TOT']<595.7) | (df1['STATUS'] == "E")]
filtro4
```

```python
#5 Solo las columnas CVE_DOC, FECHA_ENT, FECHA_VEN, CAN_TOT
filtro5 = df1[["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
filtro5
```

```python
#6 Solo las filas 7001 a 7099
filtro6 = df1.iloc[7001:7100, :]
filtro6
```

```python
#7 Index = CVE_VEND (valores 1,2) (columna = FECHAELAB)
df1 =pd.read_excel('Datos de Facturaci√≥n.xlsx', index_col= "CVE_VEND")
#df1.head()
filtro7 = df1.loc[[1, 2], ["FECHAELAB"]]
filtro7
```

```python
#convertir archivo filtrado a csv
filtro1.to_csv("Filtro1.csv")
filtro2.to_csv("Filtro2.csv")
filtro3.to_csv("Filtro3.csv")
filtro4.to_csv("Filtro4.csv")
filtro5.to_csv("Filtro5.csv")
filtro6.to_csv("Filtro6.csv")
filtro7.to_csv("Filtro7.csv")
```
