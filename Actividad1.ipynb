```python
#carga desde un archivo .xls sin indice
df =pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df.head()
```

```python
df.info()
```

## FILTROS BÁSICOS

```python
#filtro por objeto
filtro1 = df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#filtro por filas

#filtro2 = df.iloc[3:10, :] # Filas consecutivas
filtro2 = df.iloc[ [15,520,1000] , : ] #filas no consecutivas
filtro2
```

```python
#Filtro por columnas
filtro3 = df.iloc[ : , 1:10] #Consecuticas
#filtro4 = df.iloc[ : , [1,4,5]] #No consecutivas

filtro3
```

Filtro columna vs filas.
Necesitamos un índice

```python
#carga desde un archivo .xls sin indice
df1 =pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col = 1)
```

```python
#Paso 2 : Aplicamos fultro de filas vs columnas
filtro4 = df1.loc[["2022-12-23", "2022-08-25"], ["CANT"]]
filtro4
```

```python
#filtro por cabecera
filtro5 = df1.head(10)
filtro5
```

##**FILTROS BÁSICOS PARA VARIABLES CUANTITATIVAS**

```python
filtro6 =df[df["PRECIO_UNITARIO"] > 3000]
filtro6
```

##FILTRO POR CONECTORES

```python
#filtro y
filtro7 = df[(df['MARGEN_TOTAL_CALCULADO'] > 1000) & (df["NOMBRE_CLIENTE"] == "START SERVICE")]
filtro7
#7
```

```python
#filtro o (or)
filtro8 = df[(df['COSTO_TOTAL_CALCULADO'] < 200) | (df["MARGEN_TOTAL_CALCULADO"] > 10000)]
filtro8
```

```python
#filtro not
filtro9 = df1[~(df1['PRECIO_UNITARIO'] >= 500) & ~(df1["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro9
```

```python
#convertir archivo filtrado a csv
filtro1.to_csv("Filtro1.csv")
filtro2.to_csv("Filtro2.csv")
filtro3.to_csv("Filtro3.csv")
filtro4.to_csv("Filtro4.csv")
filtro5.to_csv("Filtro5.csv")
filtro6.to_csv("Filtro6.csv")
filtro7.to_csv("Filtro7.csv")
filtro8.to_csv("Filtro8.csv")
filtro9.to_csv("Filtro9.csv")
```
