```python
df =pd.read_excel('Inmuebles.xlsx', header = 1)
df.head()
```

```python
#a) Eliminar la fila 0 de unnamed y columna 0 y la columna referencia
# ELiminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
filtroA = df.drop(['Referencia'], axis=1)
filtroA
```

```python
# todos los registros de "Fecha Alta" igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00
filtroB = df[(df['Fecha_Alta']== "2007-04-29 00:00:00") | (df['Fecha_Alta']== " 2004-01-02 00:00:00")]
filtroB
```

```python
# Todos los registros de Tipo: parking, local, industrial, piso
filtroC = df[(df['Tipo']== "Parking")|(df['Tipo']== "Local")|(df['Tipo']== "Industrial")|(df['Tipo']== "Piso")]
filtroC
```

```python
#D: Todos los registros con Provincia igual a Girona y con Superficie mayor a 100
filtroD = df[(df['Provincia']=="Girona") & (df['Superficie']>100)]
filtroD
```

```python
#E: Todos los registros con precio venta: mayor a 1,000,000 o vendedor igual a "Carmen o Joaquin"
filtroE = df[(df['Precio Venta'] > 1000000) | (df['Vendedor']=="Carmen") | (df['Vendedor']=="Joaquín")]
filtroE
```

```python
#f) todos los registros que no sean de Provincia Barcenola
filtroF = df[~(df['Provincia']=="Barcelona")]
filtroF
```

```python
#g) todos los registros que no sean de "Operación": alquiler, ni tipo:oficina
filtroG = df[~(df['Operación'] == "Alquiler") & ~(df['Tipo']=="Oficina")]
filtroG
```

```python
#convertir archivo filtrado a csv
filtroA.to_csv("FiltroA.csv")
filtroB.to_csv("FiltroB.csv")
filtroC.to_csv("FiltroC.csv")
filtroD.to_csv("FiltroD.csv")
filtroE.to_csv("FiltroE.csv")
filtroF.to_csv("FiltroF.csv")
filtroG.to_csv("FiltroG.csv")
```
